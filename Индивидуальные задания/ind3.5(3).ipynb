 задача из области статистики - анализ продаж по различным регионам
  Предположим, что у нас есть данные о продажах продуктов в трех регионах - Северном, Южном и Западном.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
colors = ['orange', 'yellowgreen', 'lightskyblue']
region_sales = {'North': 2100, 'South': 1500, 'West': 1800}
plt.figure(figsize=(8,4))
plt.pie(region_sales.values(), labels=region_sales.keys(), colors=colors, autopct='%1.1f%%')
plt.title('Продажи продуктов по регионам')
plt.legend(loc='upper left')
plt.show()
```
