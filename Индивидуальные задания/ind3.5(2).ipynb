#### Задача
задача из области маркетинга - анализ продаж товаров по категориям
мы можем создать данные, которые будут служить основой для нашей столбчатой диаграммы. Предположим, что у нас есть данные о продажах товаров по 3 категориям - А, В и С, и общие продажи за последние 3 месяца:

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# данные о продажах товаров по категориям
category_sales = {'Category A': 600, 'Category B': 900, 'Category C': 1500}
# общие продажи за последние 3 месяца
total_sales = 3000
```

```python
plt.figure(figsize=(8,4))
plt.bar(category_sales.keys(), category_sales.values())
plt.xlabel('Категория товара')
plt.ylabel('Объем продаж, шт.')
plt.title('Продажи товаров по категориям')
plt.show()
```

```python
# добавление долевого соотношения продаж категорий к общим продажам
for category in category_sales:
    category_sales[category] = (category_sales[category]/total_sales) * 100
    category_sales[category] = f'{category_sales[category]:.2f}%'
```

```python
plt.figure(figsize=(8,4))
plt.bar(category_sales.keys(), category_sales.values())
plt.xlabel('Категория товара')
plt.ylabel('Доля продаж, %')
plt.title('Продажи товаров по категориям')
plt.show()
```
