#### Задача
Вычисление свободного падения тела на Земле c высоты 500 метров. Для этого нам понадобится только базовая математика и графики.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
g = 9.8
t = np.linspace(0, 10, 50) # временной отрезок от 0 до 10 секунд с 50 равномерно распределенными значениями
h = 500 - (0.5 * g * t**2) # формула для расчета высоты падающего тела на каждый момент времени
v = g * t # формула для расчета скорости падающего тела на каждый момент времени
```

Теперь, когда мы вычислили высоту и скорость падающего тела для каждого момента времени, мы можем построить графики.

```python
# построение графика высоты падающего тела
plt.plot(t, h, label='Высота')
plt.xlabel('Время, сек')
plt.ylabel('Высота, м')
plt.title('Свободное падение на Земле')
plt.legend()
plt.show()
```

```python
#построение графика скорости падающего тела
plt.plot(t, v, label='Скорость')
plt.xlabel('Время, сек')
plt.ylabel('Скорость, м/с')
plt.title('Свободное падение на Земле')
plt.legend()
plt.show()
```
